## What is RelTrans?

RelTrans is a relativistic transfer-function model used to compute X-ray spectra
and timing properties of accreting black hole systems.

Physically, it models:
- A compact X-ray source (lamppost geometry)
- General relativistic effects (redshift, lensing)
- Reflection from an accretion disk
- Spectral and timing observables (lags, covariance, etc.)

Technically:
- Core implementation: **Fortran**
- Interface layer: **Fortran wrapper subroutines**
- Execution environments:
  - XSPEC
  - Python (via shared libraries)


## Why RelTrans Cannot Be Imported Like a Python Package

RelTrans is **not** a native Python library.

It is written in Fortran and compiled into a **shared library**
(e.g. `libreltrans.so`). Python interacts with this shared library
through wrapper interfaces.

Therefore, the following will **not** work:

## Wrapper Subroutines in RelTrans

RelTrans exposes several **wrapper subroutines** designed to be called
from external environments such as XSPEC or Python.

These wrappers are defined in the file:




Examples include:
- `tdreltransPL`   – power-law–like continuum
- `tdreltransDCp`  – nthcomp continuum with density
- `tdreltransx`    – reflionx-based model
- `tdreltransDbl`  – double lamppost geometry

Each wrapper:
- Accepts simple numerical arrays
- Maps parameters internally
- Calls the core routine `genreltrans`

This design makes the code suitable for Python access.
